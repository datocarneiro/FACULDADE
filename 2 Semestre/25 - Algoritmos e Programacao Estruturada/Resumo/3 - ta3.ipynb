# ***Função malloc()***
* Alocar memória dinamicamente;
* Exemplo: 

        int* memoria = malloc (100);
* Retorna dois valores: NULL(deu errrado) ou um PONTEIRO GENÉRICO;


```python
EXEMPLO:

#include <stdio.h>
#include <stdlib.h> // Para a função rand e malloc

int* gerarRandomico() {
    int* r = (int*)malloc(10 * sizeof(int)); // Aloca memória dinamicamente
    int a;
    
    if (r != NULL) { // Verifica se a alocação de memória foi bem-sucedida
        for (a = 0; a < 10; ++a) {
            r[a] = rand();
            printf("r[%d] %d\n", a, r[a]);
        }
    } else {
        // Lida com a falha na alocação de memória
        printf("Erro na alocação de memória.\n");
    }
    
    return r;
}

int main() {
    int* randomNumbers = gerarRandomico();
    
    // Use os números aleatórios conforme necessário
    
    free(randomNumbers); // Libera a memória alocada
    
    return 0;
}
```

# ***FUNÇÕES - ESCOPO E PASSAGEM DE PARÂMENTROS***

# **_Escopo_:**

# Variável Global: 
* Criá-se fora da função, assim ela
será visível por todas as funções do programa.
Geralmente adota-se declará-las após as
bibliotecas.

# Variável Local: 
* Cria-se dentro da função, assim são “enxergadas” somente dentro
do corpo da função onde foram definidas; então logo apos o término da função o valor é apagado da memória


EXEMPLO:

```python


#include <stdio.h>

# Variável global chamada x
int x = 10; 

int testar() {
   # Acessa a variável global x
    return 2 * x; 
}
int main() {
    printf("\n=====================================================");
    # Imprime o valor da variável local x na função main
    printf("\n Valor de x global = %d", x); 
    printf("\n=====================================================");
    # Chama a função testar() que usa a variável global x
    printf("\n Valor de x global ALTERADO NA FUNÇÂO TESTAR() = %d", testar()); 
    printf("\n=====================================================");
    return 0;
}
```

# ***FUNÇÕES COM PASSAGEM DE PARÂMETROS***



# Parâmentros, Passagem por Valor:
* a função cria variáveis locais
automaticamente para armazenar esses valores e
após a execução da função essas variáveis são
liberadas.

Ideal  para ser acessada por outra função, o valor da variavel ("pai") não será alterada , só será alterada dentro da função que chamar ela de acordo com seus comando, mas ao finalisar a função atuala a variavel principal volta a ter seu valor original como antes


```python
#include <stdio.h>

int somar(int a, int b) {
    # Função somar() recebe dois parâmetros inteiros, a e b, e retorna a soma deles
    return a + b; 
}

int main() {
    int result;
    # Chama a função somar() e passa os valores 10 e 15 como argumentos para os parâmetros a e b
    result = somar(10, 15); 

    printf("\n Resultado da soma é: %d", result); 

    return 0;
}
```

# Parâmentros, passagem por referência
Se alterada em qualquer parte do codigo, o valor da varialvel original (pai) será alterada, e passa ater um novo valor 

* Ponteiro e endereço de memória;
* Não será criada uma cópia dos argumentos
passados;
* Será passado o endereço da variável e função
trabalhará diretamente com os valores ali
armazenados;

Para passar valores por referência em C, você precisa usar ponteiros. Aqui está um exemplo de como você pode modificar seu código para realizar uma passagem de valor por referência:

```python
#include <stdio.h>

void somar(int a, int b, int *resultado) {
    # A função modifica o valor apontado por 'resultado' para armazenar a soma.
    *resultado = a + b; 
}

int main() {
    int result;
    # Passa o endereço da variável 'result' para a função somar()
    somar(10, 15, &result); 
    printf("\n Resultado da soma é: %d", result);
    return 0;
}
```
