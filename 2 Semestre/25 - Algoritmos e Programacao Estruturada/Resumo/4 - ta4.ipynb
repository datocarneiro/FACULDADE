# ***PILHA*** - LIFO

São estruturas de dados do tipo LIFO (last-in first-
out), onde o último elemento a ser inserido, será o 
primeiro a ser re1rado.



Duas operações básicas:
*   Empilhar um elemento (push())
*   Desempilhar um elemento (pop())
Fonte: Ricardi (2018)

# ***FILA*** - FIFO 

São estruturas de dados do 1po FIFO (first-infirst-
out), onde o primeiro elemento a ser inserido, será
o primeiro a ser re1rado, ou seja, adiciona-se itens
no fim e remove-se do início.

Passos para a criação de uma Fila:
* criar uma fila vazia;
* inserir elemento no final;
* re1rar um elemento do início; 
* verificar se a fila está vazia;
