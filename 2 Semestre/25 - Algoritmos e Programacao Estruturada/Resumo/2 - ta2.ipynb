```python
#include <stdio.h>

int main(void) {
    int nota [5] = {11,22,33,44,55}; // [5] é quantidade de bloco dentro do vetor

    for (int i=0; i < 5; ++i ) {
        printf("%d\n", nota[i]);   
    }
        
    return 0;
}
```

```python
#include <stdio.h>

int main(void) {
    float notas [5] = {0}; 
    float total =  0;
    float media = 0;

    printf("insira as 5 notas:\n");
    for (int i=0; i < 5 ; ++i) {
        scanf("%f", &notas[i]);
    }
    
    for (int i=0; i < 5 ; ++i) {
        total += notas[i]; // soma total + nota 
    }

    media = total / 5;

    printf("A média é %f", media);
        
    return 0;
}
```
