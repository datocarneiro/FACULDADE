
# Aplicao debanco de dados com Python

Os sistema de banco de dados podem ser divididos em 
duas categorias: 
1. banco de dados relacional  (tipo uma tabela)
2. banco de dados NoSQL 

# Banco de dados relacional
Na teoria base dos bancos de dados relacionais, os dados so 
persistidos em uma estrutura bidimensional, chamada de relao 
(que  uma tabela), que est baseada na teoria dos conjuntos 
pertencentes  matemtica. Cada unidade de dados  conhecida 
como coluna, ao passo que cada unidade do grupo  conhecida 
como linha, tupla ou registro



# banco de dados No relacionais -  NoSQL 
 usado para abordar a classe de bancos de 
dados que no seguem os princpios do sistema de 
gerenciamento de banco de dados relacional (RDBMS) e 
so projetados especificamente para lidar com a 
velocidade e a escala de aplicaes



As instrues da linguagem SQL so divididas em trs 
grupos: 
1. DDL
2. DML
3. DCL

Introduo a banco de dados

1. DDL (Data Definition Language). Fazem parte deste grupo as 
instrues destinadas a criar, deletar e modificar banco de dados e 
tabelas. Neste mdulo vo aparecer comandos como CREATE, o 
ALTER e o DROP.

2. DML (Data Manipulation Language). Fazem parte deste grupo as 
instrues destinadas a recuperar, atualizar, adicionar ou excluir 
dados em um banco de dados. Neste mdulo vo aparecer 
comandos como INSERT, UPDATE e DELETE.

3. DCL (Data Control Language). Fazem parte deste grupo as 
instrues destinadas a manter a segurana adequada para o banco 
de dados. Neste mdulo vo aparecer comandos como GRANT e 
REVOK
