# Metodologia Àgil - Extreme programming  ( XP ) 
Valores: cominucação, simplicidade, feedback, coragem aberto a mudanças

** Equipe de Trabalho:
- Gerente de proejeto ( assuntos administrativos do projeto e do relacionamento com o cliente)
- Coach ( responsável técnico pelo projeto, sugerir novas tecnologias para o priejto)
- Analista de teste ( ajuda o cliente a escrever os testes de aceitação)
- Redator técnico ( ajuda a equipe de desenvolvimento a documentar o sistema)
- Desenvolvedor ( realiza análise, ajuda a criar o projeto e codifica o sistema)

* rápido desenvolvimento 
* requisitos se alteram contantemente
* protótipo
* focado em teste oe depois desenvolve
* programação em pares 



# Metodologia Agil - Scrum

• Focado na gestão de projetos, tem como base o
planejamento iterativo e incremental.

e desenvolvimento ágil de software

** Equipe de trabalho:
- Scrum Master (facilitador)
- Product Owner (responsavel pelo projeto)
- Screum Team (equipe de desenvolvimento 6 a 10 pessoas)

** Elementos:
- Produto Backlog (lista de funcionalidades)
- Sprint Backlog ( lista de tarefas a execultar naquela sprint)
- Sprint (divide o preocessso de construção em ciclos menores)


* gestão e planejamento de software
* Reuniões regulares
* Projetos são divididos em ciclos (sprints)

# GESTÃO DE CONFIGURAÇÃO DE SOFTWARE (GCS)

“É a Arte de identificar, organizar e controlar modificações no
software que está sendo criado, maximizando a
produtividade e reduzindo os erros” [Pressman, 2011]


# Planejamento de Gerenciamento de Configuração
* Definir o que será gerenciado e o esquema que será usado para identificar as entidades;
* Estabelecer o responsável;
* Estabelecer políticas de gerenciamento de configuração;
* Especificar de ferramentas;
* Descrever a estrutura do banco de dados (ierarquia)


# 5 Tarefas para se ter uma boa gestão de configuração de software

De dentro para fora (de baixo para cima)



# * Repositório / Baselines /  Branches;

* Repositório
    * Local onde colocamos os nossos programas em desenvolvimentos e demias arquivos (armazenamento)
- Baseline (linha de base)
    - É meu conjuntos de itens de configuração (base principal) formalmente aprovavedos server de base para as proximas etapas dando segurança no desenvolvimento, porem quando o programa é entregue ao cliente é chamado de release
* Branches
    * São as ramificações, exemplo quando vamos incluir novas funcionalidades no programa, ou alteração, não são feitas na baseline e sim nas branches, garantido o código base estável


# O que são Item de configuração?
Elemento unitário ou um grupo de elementos para efeito de controle de versão. 
* Código
* Documentação
* Diagramas, planos, ferramentas, casos de teste e etc


# QUal a importancia da Identificação de item de configuração?
* Manter a rastreabilidade (encontrar o que precisa quando necessário, muito util para manutenção, identificação clara , padrão entre a equipe)
* Documentos que podem ser úteis devem ser controlados pelo gerenciamento de configuração;
* Esquemas de nomes hierarquizados;


# Banco de Dados de Configuração (onde salvar, salvar legendas etc)
É um repositório salvando as informações que eu preciso 
* Registrar todas as informações relevantes sobre a configurações de sistemas e os itens de configuração (seus detalhes);
* Avaliar o impacto das mudanças no sistema;
* Registro e recuperação de informações de projeto (ser acessivel quando necessario);


# Gerenciamento de Versões e Releases
* Preocupa-se com a identificação e a rastreabilidade das versões de um sistema;
* Uma versão é uma instância de um sistema, que difere, de alguma maneira, de outras instâncias;
* Versões com pequenas diferenças são algumas vezes chamadas de variantes;

- RELEASES:
    - Versão distribuída aos clientes;
        - Os releases devem incorporar novas funcionalidades ou ser planejado para uma plataforma diferente de hardware;
            - Há normalmente muito mais versões de um sistema do que liberações;



# Identificação de Versões 
* Numeração de versões: exemplo: 1.1, 2.1 ...
* Identificação baseada em atributos;
* Identificação orientada a mudanças;
