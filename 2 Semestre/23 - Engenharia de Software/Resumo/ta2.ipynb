# Custo da Qualidade 



# Um levantamento de requisitos bem feito  impacta diretamente na  Qualidade e no custo futuros
- Requisitos Funcionais 

    Exemplos:
    - Incluir/Excluir/Alterar nome em uma tela de manutenção de funcionário
    - Geração de relatório de determinado período de vendas
    - Efetuar pagamentos de compra através de crédito ou débito
    - Consulta de saldo ou estoque
    
- Requisitos Não Funcionais

    Exemplos:
    - O sistema deverá ter alta disponibilidade
    - O sistema deverá executar em qualquer plataforma.
    - O sistema deverá se comunicar com o banco SQL Server.




# SQA - Garantia da Qualidade 
* A SQA prega que em todas as fases do desenvolvimento podemos ter ferramentas que garantem a qualidade 
    - avaliação
    - analise 
    - teste
    - manutenção 

- SQA Deve:
    - Possuir ferramentas e/ou métodos que permitam a análise dos desenvolvimentos e dos testes.
    - Efetuar revisões técnicas nos componentes e nafuncionalidade;
    - Controlar a documentação por meio de versionamento. (a documentação precisa caminhar junto com as atualizações de versão do código)
    - Atribuir métodos para se garantir padrões de desenvolvimento e das boas práticas,
    - Obter mecanismos de aferição.

# *** ISO 9000 ***
* Descrever os fundamentos e princípios da gestão da qualidade.
* Compreender os processos de implementação da gestão da qualidade.
* Avaliar a conformidade dos produtos de software desenvolvido.

Prega: 

* Foco no cliente: 
    * uma abordagem por meio da qual se buscammelhores práticas, a fim de entregar o melhor produto.
* Liderança: 
    * metodologia e abordagens como forma de liderar.
* Pessoas: 
    *  utilizar formas de as pessoas se comprometerem com osprocessos e com a qualidade.
* Processos: 
    * verificar constantemente os processos e repensá-los.
* Inter-relacionamento: 
    * prover o inter-relacionamento deatividades concorrentes.
* Melhoria: 
    * buscar a melhoria contínua por meio de metodologias, normas e boas práticas.
* Decisão:
    * utilizar os feedbacks gerados a favor da tomada dedecisão.
* Benefícios: 
    * gerar vantagens administrativas e operacionais por meio da adoção de boas práticas.

# *** QUALIDADE DO PROCESSO – CMMI e MPS.BR ***

Os dois modelos mais conhecido são 
* CMMI
* MPS.BR

funcionam da seguinte maneira, com nivies de maturidade, quanto a empresa consegue entregar do que é proposto, recebem um selo de acordo com a avaliação de maturidade realizada pelo auditor

# CMMI
É um modelo de referência que aborda a melhoria deprocessos.
* Dividido em 5 níveis de maturidade;

    

    * Três modelos:
    * CMMI for Development (CMMI-DEV) - para desenvolvimeneto 
    * CMMI for Acquisition (CMMI-ACQ) - Aquisição
    * CMMI for Services (CMMI-SVC) - Serviços

# CMMI - Contém 22 áreas de processos

# As áreas processos são divididas em 4 grupos:
    * Gerenciamento de processos
    * Gerenciamento de projetos
    * Engenharia
    * Apoio

* Gerenciamento de processos:
    * Foco no processo organizacional;
    * Definição do processo;
    * Treinamento;
    * Desempenho do processo organizacional;
* Gerenciamento de Projetos:
    * Planejamento de projetos;
    * Monitoração e controle de projeto;
    * Fornecedores;
    * Gerenciamento de Risco;
    * Gerenciamento de projetos integrados;
* Engenharia:
    * Gerenciamento de requisitos
    * Desenvolvimento de requisitos
    * Solução técnica
    * Integração de produto (junção de codigos)
    * Verificação (testes)
    * Validação (testes)
* Apoio:
    * Gerenciamento de configuração (repositorio, nomenclaturas padrão)
    * Gerenciamento de qualidade de processo e produto; 
    * Medição e análise ;
    * Análise de decisão e resolução ;
    * Análise causal e resolução; 




# MPS.BR (padrão brasileiro)
Criado em 2003 pela Softex.

* Sete níveis de Maturidade. (contra 5 do CMMI)
* Compatível com o CMMI.
* Três modelos de referências:
    * MPS-SW - (desonvolvimento de software)
    * MPS-SV - (Serviços)
    * MPS-RH - (recursos Humanos)

MPS.BR - vem crescedo ao longo do tempo:

* Processo Organizacionais
