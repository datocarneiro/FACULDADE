# ***Processo  de negócio***
Nada mais é que uma sequencia lógica estrutura de tarefas que é uma entrada um processo  e uma saida

complexidade: 

         existem simples e complexos 

# Fundamentos:

        ENTRADA -> OPERAÇÕES OU PROCESSAMENTOS -> SAÍDAS OU RESULTADOS 

        podem ser retroalimentadas ou não 


è importante ter uma visão geral de todo o precesso de negócio

assim nós como analista conseguimos estudar onde o sistema vai estar inserido, quais funcionalidades do sistema vão estar envolvidos para cada departamento por exemplo 

e para se ter uma boa visão do processo de negócio usamos a modelagem de sistema BPMN

# ***BPMN - Business Process Modeling Notation***

Nada mais é como se fosse um tipo de fluxograma, com a sua notação especifica, entendendo como uma empresa funciona no geral.

é uma modelagem que fazemos envolvendo toda a empresa 

quais departamentos o sistema que vamos desenvolver, vai ter integração quias as funcionalidades
