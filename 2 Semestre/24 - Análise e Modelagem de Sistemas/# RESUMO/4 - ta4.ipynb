# ***Paradigma Orientado a Objetos (POO)***
* È um Padrão de desenvolvimento;
    * modela usando modelos do mundo real
* Como modelar os problemas do mundo real;
* Smaltalk 80;
    * foi criado na década de 60  em (1966), mas to tomou forma mesmo nos anos 80 com a smalltalk 80
* C++;
    * C++ foi quem deu origem (++, significa if class )  


# * Conceitos básicos da POO* 

*  Abstração; 
    - idéia central
*  Classe; 
    - representa as caracteristicas
*  Atributos e métodos; 
    - ações 
*  Objeto; 
    - materialização de forma única da classe
*  Herança; 
    - classes que derivam de outras classes (tipo Pai, filho)
*  Encapsulamento; 
    - proteção dos atributos (permissão)
*  Polimorfismo; 
    - objeto do  mesmo tipo com comportamentos diferentes

# ***Diagrama de classes - UML***

# 1ºPARTE 
No diagrama de classe usamos o:

        símbolo “-“ (um traço) para indicar que o atributo ou o método está encapsulado. Isto significa que somente o objeto da classe pode modificá-los ou acessá-los. 

        Já o símbolo “+” (adição) indica que o atributo ou método é visível por outros objetos, ou seja, é público. 

        E o símbolo "#" significa que as classes de herançã alem da classa PAI a de herança tabm podem visualizar e acessar
