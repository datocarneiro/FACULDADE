# ***CASOS DE USO***

# Componentes de caso de uso

* Ator

         usuario ou outro sistema 

* casos de uso

        onde faz as especificações de requisitos, com títuo/frase pequeno(a), e a primeira palavra usamos um verbo, algo que demonstre ação

* Limite do sistema

        delimita quem pode utilizar o sistema

* Associoação de comunicação

        ligação entre atores e casos de uso

* include

        só acontece entre casos de uso relacionamento que impoem obrigatoriadade de acontecer 

* Extend

        Entre casos de uso, para comportamentos opcionais

* Generalização

        utiliza para fazer uma herança
