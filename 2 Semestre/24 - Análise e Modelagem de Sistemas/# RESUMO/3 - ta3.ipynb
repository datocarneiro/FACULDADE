# ***CASOS DE USO***

# Componentes de caso de uso

* Ator

         usuario ou outro sistema 

* casos de uso

        onde faz as especificaes de requisitos, com ttuo/frase pequeno(a), e a primeira palavra usamos um verbo, algo que demonstre ao

* Limite do sistema

        delimita quem pode utilizar o sistema

* Associoao de comunicao

        ligao entre atores e casos de uso

* include

        s acontece entre casos de uso relacionamento que impoem obrigatoriadade de acontecer 

* Extend

        Entre casos de uso, para comportamentos opcionais

* Generalizao

        utiliza para fazer uma herana
